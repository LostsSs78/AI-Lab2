Пример 21 Умножение матрицы на число

```python
A = np.matrix('1 2 3; 4 5 6')
C = 3 * A
print(C)
print("-"*20)

# Умножение на 1 даёт исходную матрицу
A = np.matrix('1 2; 3 4')
L = 1 * A
R = A
print(L)
print(R)
print("-"*20)

# Умножение на 0 даёт нулевую матрицу с размерностью исходной
A = np.matrix('1 2; 3 4')
Z = np.matrix('0 0; 0 0')
L = 0 * A
R = Z
print(L)
print(R)
print("-"*20)

# Свойство 3. Произведение матрицы на сумму чисел равно сумме 
# произведений матрицы на каждое из этих чисел:

A = np.matrix('1 2; 3 4')
p = 2
q = 3
L = (p + q) * A
R = p * A + q * A
print(L)
print(R)
print("-"*20)

# Свойство 4. Произведение матрицы на произведение двух 
# чисел равно произведению второго числа и заданной матрицы, 
# умноженному на первое число:

A = np.matrix('1 2; 3 4')
p = 2
q = 3
L = (p * q) * A
R = p * (q * A)
print(L)
print(R)
print("-"*20)

# Свойство 5. Произведение суммы матриц на число равно 
# сумме произведений этих матриц на заданное число:

A = np.matrix('1 2; 3 4')
B = np.matrix('5 6; 7 8')
k = 3
L = k * (A + B)
R = k * A + k * B
print(L)
print(R)
```

Пример 22 Сложение матриц

```python
A = np.matrix('1 6 3; 8 2 7')
B = np.matrix('8 1 5; 6 9 12')
C = A + B
print(C)
print("-"*20)

# Свойство 1. Коммутативность сложения. 
# От перестановки матриц их сумма не изменяется:
# A + B = B + A.

A = np.matrix('1 2; 3 4')
B = np.matrix('5 6; 7 8')
L = A + B
R = B + A
print(L)
print(R)
print("-"*20)

# Свойство 2. Ассоциативность сложения. 
# Результат сложения трех и более матриц не зависит от порядка, 
# в котором эта операция будет выполняться:
# A + (B + C) = (A + B) + C.

A = np.matrix('1 2; 3 4')
B = np.matrix('5 6; 7 8')
C = np.matrix('1 7; 9 3')
L = A + (B + C)
R = (A + B) + C
print(L)
print(R)
print("-"*20)

# Свойство 3. Для любой матрицы существует противоположная ей, 
# такая, что их сумма является нулевой матрицей:
# A + (-A) = Z.

A = np.matrix('1 2; 3 4')
Z = np.matrix('0 0; 0 0')
L = A + (-1) * A
print(L)
print(Z)
```

Пример 23 Умножение матриц

```python
A = np.matrix('1 2 3; 4 5 6')
B = np.matrix('7 8; 9 1; 2 3')
C = A.dot(B)
print(C)
print("-"*20)

# Свойство 1. Ассоциативность умножения. 
# Результат умножения матриц не зависит от порядка, 
# в котором будет выполняться эта операция:
# A × (B × C) = (A × B) × C.

A = np.matrix('1 2; 3 4')
B = np.matrix('5 6; 7 8')
C = np.matrix('2 4; 7 8')
L = A.dot(B.dot(C))
R = (A.dot(B)).dot(C)
print(L)
print(R)
print("-"*20)

# Свойство 2. Дистрибутивность умножения. 
# Произведение матрицы на сумму матриц равно сумме 
# произведений матриц:
# A × (B + C) = A × B + A × C.

A = np.matrix('1 2; 3 4')
B = np.matrix('5 6; 7 8')
C = np.matrix('2 4; 7 8')
L = A.dot(B + C)
R = A.dot(B) + A.dot(C)
print(L)
print(R)
print("-"*20)

# Свойство 3. Умножение матриц в общем виде не коммутативно. 
# Это означает, что для матриц не выполняется правило независимости 
# произведения от перестановки множителей:
# A × B ≠ B × A.

A = np.matrix('1 2; 3 4')
B = np.matrix('5 6; 7 8')
L = A.dot(B)
R = B.dot(A)
print(L)
print(R)
print("-"*20)

# Свойство 4. Произведение заданной матрицы на единичную равно 
# исходной матрице:
# E × A = A × E = A.

A = np.matrix('1 2; 3 4')
E = np.matrix('1 0; 0 1')
L = E.dot(A)
R = A.dot(E)
print(L)
print(R)
print(A)
print("-"*20)

# Свойство 5. Произведение заданной матрицы на нулевую матрицу 
# равно нулевой матрице:
# Z × A = A × Z = Z.

A = np.matrix('1 2; 3 4')
Z = np.matrix('0 0; 0 0')
L = Z.dot(A)
R = A.dot(Z)
print(L)
print(R)
print(Z)
```

Пример 24 Определитель матрицы

```python
A = np.matrix('-4 -1 2; 10 4 -1; 8 3 1')
print(A)
print("-"*20)

# Для вычисления определителя этой матрицы воспользуемся функцией 
# det() из пакета linalg.

print(np.linalg.det(A))
print("-"*20)

# Свойство 1. Определитель матрицы остается неизменным 
# при ее транспонировании:
# det(A) = det(A^T).

print("Свойство 1")
A = np.matrix('-4 -1 2; 10 4 -1; 8 3 1')
print(A)
print(A.T)
print("-"*20)

det_A = round(np.linalg.det(A), 3)
det_A_t = round(np.linalg.det(A.T), 3)
print(det_A)
print(det_A_t)
print("-"*20)

# Свойство 2. Если у матрицы есть строка или столбец, 
# состоящие из нулей, то определитель такой матрицы равен нулю.

print("Свойство 2")
A = np.matrix('-4 -1 2; 0 0 0; 8 3 1')
print(A)
print(np.linalg.det(A))
print("-"*20)

# Свойство 3. При перестановке строк матрицы знак 
# ее определителя меняется на противоположный.

print("Свойство 3")
A = np.matrix('-4 -1 2; 10 4 -1; 8 3 1')
print(A)

B = np.matrix('10 4 -1; -4 -1 2; 8 3 1')
print(B)

print(round(np.linalg.det(A), 3))
print(round(np.linalg.det(B), 3))
print("-"*20)

# Свойство 4. Если у матрицы есть две одинаковые строки, 
# то ее определитель равен нулю.

print("Свойство 4")
A = np.matrix('-4 -1 2; -4 -1 2; 8 3 1')
print(A)
print(np.linalg.det(A))
print("-"*20)

# Свойство 5. Если все элементы строки или столбца матрицы 
# умножить на какое-то число, то и определитель будет умножен 
# на это число.

print("Свойство 5")
A = np.matrix('-4 -1 2; 10 4 -1; 8 3 1')
print(A)
k = 2
B = A.copy()
B[2, :] = k * B[2, :]
print(B)
det_A = round(np.linalg.det(A), 3)
det_B = round(np.linalg.det(B), 3)
print(det_A * k)
print(det_B)
print("-"*20)

# Свойство 6.

print("Свойство 6")
A = np.matrix('-4 -1 2; -4 -1 2; 8 3 1')
B = np.matrix('-4 -1 2; 8 3 1; 8 3 1')
print(round(np.linalg.det(A), 3) + round(np.linalg.det(B), 3))
print("-"*20)

# Свойство 7. Если к элементам одной строки прибавить элементы 
# другой строки, умноженные на одно и тоже число, то определитель 
# матрицы не изменится.

print("Свойство 7")
A = np.matrix('-4 -1 2; 10 4 -1; 8 3 1')
k = 2
B = A.copy()
B[1, :] = B[1, :] + k * B[0, :]
print(A)
print(B)
print(round(np.linalg.det(A), 3))
print(round(np.linalg.det(B), 3))
print("-"*20)

# Свойство 8. Если строка или столбец матрицы является 
# линейной комбинацией других строк (столбцов), то определитель 
# такой матрицы равен нулю.

print("Свойство 8")
A = np.matrix('-4 -1 2; 10 4 -1; 8 3 1')
print(A)
k = 2
A[1, :] = A[0, :] + k * A[2, :]
print(round(np.linalg.det(A), 3))
print("-"*20)

# Свойство 9. Если матрица содержит пропорциональные строки, 
# то ее определитель равен нулю.

print("Свойство 9")
A = np.matrix('-4 -1 2; 10 4 -1; 8 3 1')
print(A)
k = 2
A[1, :] = k * A[0, :]
print(A)
print(round(np.linalg.det(A), 3))
```

Пример 25 Обратная матрица

```python
A = np.matrix('1 -3; 2 5')
A_inv = np.linalg.inv(A)
print(A_inv)
print("-"*20)

# Свойство 1. Обратная матрица обратной матрицы есть 
# исходная матрица:
# (A^(-1))^(-1) = A.

A = np.matrix('1. -3.; 2. 5.')
A_inv = np.linalg.inv(A)
A_inv_inv = np.linalg.inv(A_inv)
print(A)
print(A_inv_inv)
print("-"*20)

# Свойство 2. Обратная матрица транспонированной матрицы 
# равна транспонированной матрице от обратной матрицы:
# (A^T)^(-1) = (A^(-1))^T.

A = np.matrix('1. -3.; 2. 5.')
L = np.linalg.inv(A.T)
R = (np.linalg.inv(A)).T
print(L)
print(R)
print("-"*20)

# Свойство 3. Обратная матрица произведения матриц равна 
# произведению обратных матриц:
# (A1 × A2)^(-1) = A2^(-1) × A1^(-1).

A = np.matrix('1. -3.; 2. 5.')
B = np.matrix('7. 6.; 1. 8.')
L = np.linalg.inv(A.dot(B))
R = np.linalg.inv(B).dot(np.linalg.inv(A))
print(L)
print(R)
```

Пример 26 Ранг матрицы

```python
m_eye = np.eye(4)
print(m_eye)

# Ранг такой матрицы равен количеству ее столбцов (или строк), 
# в нашем случае ранг будет равен четырем, для его вычисления 
# на Python воспользуемся функцией matrix_rank():

rank = np.linalg.matrix_rank(m_eye)
print(rank)
print("-"*20)

# Если мы приравняем элемент в нижнем правом углу к нулю, 
# то ранг станет равен трем:

m_eye[3][3] = 0
print(m_eye)

rank = np.linalg.matrix_rank(m_eye)
print(rank)
```
