Задание 1.
Условие:
Создайте массив NumPy размером 3×3, содержащий числа от 1 до 9. Умножьте все элементы массива на 2, а затем замените все элементы больше 10 на 0. Выведите итоговый массив.

```python
import numpy as np

# Создаем массив 3x3 с числами от 1 до 9
array_3x3 = np.arange(1, 10).reshape(3, 3)
print("Исходный массив:")
print(array_3x3)

# Умножаем все элементы на 2
array_multiplied = array_3x3 * 2
print("\nМассив после умножения на 2:")
print(array_multiplied)

# Заменяем элементы больше 10 на 0
array_multiplied[array_multiplied > 10] = 0
print("\nИтоговый массив (элементы > 10 заменены на 0):")
print(array_multiplied)
```

Условие:
Создайте два массива NumPy размером 1×5, заполненные случайными числами от 0 до 50. Объедините эти массивы в один двумерный массив (по строкам). Разделите полученный массив на два массива, каждый из которых содержит 5 элементов. Выведите все промежуточные и итоговые результаты.

```python
# Создаем два массива 1x5 со случайными числами от 0 до 50
array1 = np.random.randint(0, 51, size=(1, 5))
array2 = np.random.randint(0, 51, size=(1, 5))

print("Первый массив (1x5):")
print(array1)
print("\nВторой массив (1x5):")
print(array2)

# Объединяем массивы по строкам (вертикальное объединение)
merged_array = np.vstack((array1, array2))
print("\nОбъединенный массив (по строкам):")
print(merged_array)
print(f"Форма объединенного массива: {merged_array.shape}")

# Разделяем полученный массив на два массива по 5 элементов каждый
# Вариант 1: разделение по строкам (каждая строка - отдельный массив)
split_arrays = np.vsplit(merged_array, 2)
print("\nРазделение по строкам (каждый массив содержит 5 элементов):")
print("Первый массив:")
print(split_arrays[0])
print("\nВторой массив:")
print(split_arrays[1])

# Вариант 2: разделение на два массива с формой (1, 5)
# Альтернативный способ - через reshape
first_part = merged_array[0:1, :]  # Первая строка
second_part = merged_array[1:2, :]  # Вторая строка
print("\nАльтернативный способ разделения:")
print("Первый массив (первая строка):")
print(first_part)
print("\nВторой массив (вторая строка):")
print(second_part)
```

Задание 4. 
Условие:
Создайте массив из 50 чисел, равномерно распределенных от -10 до 10. Вычислите сумму всех элементов, сумму положительных элементов и сумму отрицательных элементов. Выведите результаты.

```python
# Создаем массив из 50 чисел, равномерно распределенных от -10 до 10
linear_array = np.linspace(-10, 10, 50)
print("Массив из 50 чисел, равномерно распределенных от -10 до 10:")
print(linear_array)
print(f"\nКоличество элементов: {len(linear_array)}")
print(f"Первый элемент: {linear_array[0]:.2f}")
print(f"Последний элемент: {linear_array[-1]:.2f}")

# Вычисляем сумму всех элементов
total_sum = np.sum(linear_array)
print(f"\nСумма всех элементов: {total_sum:.4f}")

# Вычисляем сумму положительных элементов
positive_mask = linear_array > 0
positive_sum = np.sum(linear_array[positive_mask])
print(f"Сумма положительных элементов: {positive_sum:.4f}")

# Вычисляем сумму отрицательных элементов
negative_mask = linear_array < 0
negative_sum = np.sum(linear_array[negative_mask])
print(f"Сумма отрицательных элементов: {negative_sum:.4f}")
```

Задание 5. 
Условие:
Создайте единичную матрицу размером 4×4. Создайте диагональную матрицу размером 4×4 с диагональными элементами [5, 10, 15, 20] (не использовать циклы). Найдите сумму всех элементов каждой из этих матриц и сравните результаты.

```python
# Создаем единичную матрицу 4x4
identity_matrix = np.eye(4)
print("Единичная матрица 4x4:")
print(identity_matrix)

# Создаем диагональную матрицу 4x4 с диагональными элементами [5, 10, 15, 20]
diagonal_elements = np.array([5, 10, 15, 20])
diagonal_matrix = np.diag(diagonal_elements)
print("\nДиагональная матрица 4x4:")
print(diagonal_matrix)

# Находим сумму всех элементов каждой матрицы
sum_identity = np.sum(identity_matrix)
sum_diagonal = np.sum(diagonal_matrix)

print(f"\nСумма всех элементов единичной матрицы: {sum_identity}")
print(f"Сумма всех элементов диагональной матрицы: {sum_diagonal}")

# Сравниваем результаты
print(f"\nСравнение результатов:")
print(f"Сумма диагональной матрицы больше суммы единичной на {sum_diagonal - sum_identity}")
print(f"Сумма диагональной матрицы в {sum_diagonal / sum_identity:.2f} раз больше суммы единичной")
```

Задание 7. Условие:
Создайте две матрицы NumPy: первую размером 2×3, заполненную случайными числами от 1 до 10, и вторую размером 3×2, заполненную случайными числами от 1 до 10. Выполните матричное умножение (@ или np.dot) и выведите результат.

```python
# Создаем матрицу A размером 2x3 со случайными числами от 1 до 10
matrix_a = np.random.randint(1, 11, size=(2, 3))
print("Матрица A (2x3):")
print(matrix_a)
print(f"Размерность A: {matrix_a.shape}")

# Создаем матрицу B размером 3x2 со случайными числами от 1 до 10
matrix_b = np.random.randint(1, 11, size=(3, 2))
print("\nМатрица B (3x2):")
print(matrix_b)
print(f"Размерность B: {matrix_b.shape}")

# Выполняем матричное умножение (способ 1 - с помощью @)
result_at = matrix_a @ matrix_b
print("\nРезультат матричного умножения (A @ B):")
print(result_at)
print(f"Размерность результата: {result_at.shape}")

# Выполняем матричное умножение (способ 2 - с помощью np.dot)
result_dot = np.dot(matrix_a, matrix_b)
print("\nРезультат матричного умножения (np.dot(A, B)):")
print(result_dot)
print(f"Размерность результата: {result_dot.shape}")

# Проверяем, что оба способа дают одинаковый результат
print(f"\nРезультаты совпадают: {np.array_equal(result_at, result_dot)}")
```

ИНДИВИДУАЛЬНОЕ ЗАДАНИЕ. Для создания нового металлического сплава используются три вида металлов. В первом металле содержится 60% железа, во втором — 70%, в третьем — 80%. Нужно получить 100 кг сплава с 72% железа. При этом второго металла должно быть в два раза больше, чем первого. Сколько килограммов каждого металла нужно взять?

Матричный метод

```python
# Матрица коэффициентов
matrix_a = np.array([
    [1, 1, 1],
    [0.6, 0.7, 0.8],
    [-2, 1, 0]
])

# Вектор правой части
vector_b = np.array([100, 72, 0])

# Решение системы
solution_matrix = np.linalg.solve(matrix_a, vector_b)

print("Матричный метод:")
print(f"x = {solution_matrix[0]:.2f} кг")
print(f"y = {solution_matrix[1]:.2f} кг")
print(f"z = {solution_matrix[2]:.2f} кг")
```

Метод Крамера

```python
# Определитель основной матрицы
det_a = np.linalg.det(matrix_a)

# Матрицы для метода Крамера
matrix_x = matrix_a.copy()
matrix_x[:, 0] = vector_b

matrix_y = matrix_a.copy()
matrix_y[:, 1] = vector_b

matrix_z = matrix_a.copy()
matrix_z[:, 2] = vector_b

# Вычисление решений
x = np.linalg.det(matrix_x) / det_a
y = np.linalg.det(matrix_y) / det_a
z = np.linalg.det(matrix_z) / det_a

print("\nМетод Крамера:")
print(f"x = {x:.2f} кг")
print(f"y = {y:.2f} кг")
print(f"z = {z:.2f} кг")
```
